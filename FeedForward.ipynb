```python
import numpy as np
```

```python
X=np.array([[1,0,1,0],[1,0,1,1],[0,1,0,1]])
y=np.array([[1],[1],[0]])
```

```python
def sigmoid_activation(x):
    return 1/ (1+np.exp(-x))
def derivative(x):
    return x * (1-x)
```

```python
input_neurons=X.shape[1]
hidden_neurons = 3
output_neurons = 1
alpha=0.1 # learning rate
epochs=6000

# weights and bias initialization
wh = np.random.uniform(size=(input_neurons, hidden_neurons))
bh= np.random.uniform(size=(1,hidden_neurons))
wout = np.random.uniform(size=(hidden_neurons, output_neurons))
bout = np.random.uniform(size=(1, output_neurons))
```

```python
# feed forward
for i in range(epochs):
    hidden_layer_input=np.dot(X,wh) + bh
    hidden_layer_activation = sigmoid_activation(hidden_layer_input)
    output_layer=np.dot(hidden_layer_activation, wout) + bout
    output=sigmoid_activation(output_layer)
    
    # Back propagation
    output_loss=y-output
    output_slope=derivative(output)
    output_delta=output_loss*output_slope
    
    hidden_loss= output_delta.dot(wout.T)
    hidden_slope=derivative(hidden_layer_activation)
    hidden_delta=hidden_loss*hidden_slope
    
    wout=wout+hidden_layer_activation.T.dot(output_delta)*alpha
    wh=wh+X.T.dot(hidden_delta)*alpha
    bout=bout+np.sum(output_delta, axis=0, keepdims=True)
    bh=bh+np.sum(hidden_delta,axis=0, keepdims=True)*alpha
```
