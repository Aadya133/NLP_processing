#NLTK - Natural Language ToolKit

```python
def preprocess(documents):

  import nltk
  import string

  nltk.download("punkt")
  nltk.download("stopwords")
  nltk.download("wordnet")
  nltk.download("punkt_tab")

  from nltk.tokenize import word_tokenize
  from nltk.corpus import stopwords
  from nltk.stem import WordNetLemmatizer
  from nltk.stem import PorterStemmer

  processed_doc=[]
  eng_stopwords=stopwords.words("english")
  punctuations=string.punctuation
  wnet=WordNetLemmatizer()

  #for each sentence
  for text in documents :

    #lower
    text=text.lower()
    tokens=word_tokenize(text)

    #removing stopwords
    filtered_tokens=[x for x in tokens if x not in eng_stopwords]

    #remove punctuations
    clean_tokens=[x for x in filtered_tokens if x not in punctuations]
    clean_tokens=[x for x in clean_tokens if x.isalpha()]

    #lammetize
    Lemmatized_tokens=[wnet.lemmatize(x,"v") for x in clean_tokens]

    #join a sentence
    cleaned_text=" ".join(Lemmatized_tokens)

    #append to new doc
    processed_doc.append(cleaned_text)

  return processed_doc
```
