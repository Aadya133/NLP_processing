```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
```

```python
sentences =[
    "I am learning natural learning processing",
    "Natural learning processing is part of machine learning",
    "I am learning machine learning also",
    "I want to learn deep learning",
    "deep learning uses neural networks", 
    "Neural networks are powerful"
]
```

```python
tokenized_sentences =[word_tokenize(s.lower()) for s in sentences]
```

```python
tokenized_sentences
```

```python
model=Word2Vec(
    sentences=tokenized_sentences, # 
    vector_size=100, # embedding dimension - dimension of word_vectors
    window=3, # context window size - max distance between target and context words
    min_count=1 # keep all words- ignore words appearing less than min_count
)
```

```python
model.wv.similarity("machine","uses")
```

```python
model.wv.most_similar("machine", topn=5)
```

# Internally :
During training :
model slides window across text
learn weights via neural network
produces embedding matrix

Word2Vec is a shallow neural network with one hidden layer
Limitation - static embedding - same vector for every context
