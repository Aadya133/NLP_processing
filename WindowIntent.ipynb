# Window based Intent predictor


```python
# predict the intent using a sliding window over sentence embeddings
# This will make us understand that how embeddings are used as features
```

```python
import gensim.downloader as api
import numpy as np
from nltk.tokenize import word_tokenize
import nltk
```

```python
print("Loading Model")
model=api.load("glove-wiki-gigaword-100")
print("Loaded")
```

```python
sentence="Please open google and check weather"
tokens=word_tokenize(sentence.lower())
```

```python
# Sliding window function
def get_windows(tokens, window_size=2):
    windows=[]
    for i in range(len(tokens)):
        left=max(0,i-window_size)
        right=min(len(tokens), i+window_size+1)
        window_words=tokens[left:right]
        windows.append(window_words)
    return windows
```

```python
windows=get_windows(tokens)
windows
```

```python
# Convert window to embedding
def window_vector(window):
    vectors=[]
    for word in window:
        if word in model:
            # get embedding of a word from given model
            vectors.append(model[word])
    if len(vectors)==0:
        return np.zeroes(100)
    return np.mean(vectors,axis=0)
```

```python
windows_vector=[window_vector(v) for v in windows]
windows_vector[0].shape
```
