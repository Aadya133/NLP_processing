```python
import tensorflow as tr
from tensorflow.keras import datasets,layers,models
import matplotlib.pyplot as plt
```

```python
# Load dataset
(train_images,train_labels),(test_images,test_labels)=datasets.fashion_mnist.load_data()
```

```python
# Normalization of pixel values
train_images=train_images/255.0
test_images=test_images/255.0
```

```python
train_images[0].shape
```

```python
model=models.Sequential()

# Convolutional layer
# 16 filters, 3*3 kernel, activation - relu, input_shape = (28,28,1)- 1- for single colour channel
model.add(layers.Conv2D(16,(3,3),activation='relu',input_shape=(28,28,1)))
# Pooling - reduce size
model.add(layers.MaxPooling2D((2,2)))

#convolution layer
model.add(layers.Conv2D(32,(2,2),activation='relu'))
#Pooling
model.add(layers.MaxPooling2D((2,2)))

# Flatten - map 2d features to 1d vectors - at last only one time
model.add(layers.Flatten())

# Fully connected Layers
# we can use any activation here
model.add(layers.Dense(64,activation='relu'))

# Output Layer , softmax- probability
model.add(layers.Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss=tr.keras.losses.sparse_categorical_crossentropy(),metrics=['accuracy'])
model.fit(train_images,train_labels,validation_data=(test_images,test_labels),epochs=10)
```
