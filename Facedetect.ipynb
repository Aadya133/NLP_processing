# Face Detection


```python
import cv2
import numpy as np
```

```python
# Load haar cascade for face detection
haar_features=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
img=cv2.imread("face.jpg")
# output -> array 
# rows = no of face detected
# col- > coordinate of face (x,y,width,height)

#parameters 
# 1- image
# 2- scaleFactor - how much image is reduced at each step
faces = haar_features.detectMultiScale(img)

print(faces)
```

```python
for x,y,w,h in faces:
    # parameters - >( img, (x,y coordinate (top-left)), (x+width, y+height(bottom right)),(colour(B,G,R)),thickness)
    cv2.rectangle(img, (x,y), (x+w, y+h),(0,255,0),2)
```

```python
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
