```python
# Text -> Tokenisation -> Sequences -> Embedding -> LSTM -> Prediction
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
# dense -> o/p layer, hidden layer
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences 
```

```python
sentences=[
    "i really liked the movie",
    "the film was amazing",
    "i like the first part but not the second",
    "movie was terrible",
    "The movie was boring",
    "I hate that film",
    "the movie was good"
]
# 1-> positive , 0 -> negative
labels=np.array([1,1,1,0,0,0,1])
```

```python
# create object of tokenizers
tokenizer=Tokenizer()
# tokenize sentence-> words
tokenizer.fit_on_texts(sentences)
```

```python
# convert Text - >numbers
sequences=tokenizer.texts_to_sequences(sentences)
print(sequences)
```

```python
# Make all sentences of same length - > pad_sequences
X=pad_sequences(sequences)
X
```

```python
# word index - > starts from 0 => size = +1
vocab_size=len(tokenizer.word_index)+1
max_len=X.shape[1] # no of columns
```

```python
# sequential groups a linear stack of layers into a model
model=Sequential([
    # Embedding layer 
    Embedding(input_dim=vocab_size,output_dim=10,input_length=max_len),
    #LSTM layer
    LSTM(32),
    #Output Layer
    Dense(1, activation="sigmoid")
])
```

```python
# loss fn -> 
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,labels,epochs=100)
```

```python
def predict_sentiment(text):
  seq=tokenizer.texts_to_sequences({text})
  padded=pad_sequences(seq,maxlen=max_len)
  prediction=model.predict(padded)
  print(prediction)
  if prediction>0.5:
    print("Positive")
  else:
    print("Negative")
predict_sentiment("Movie was amazing")
predict_sentiment("I really liked the movie")
```
