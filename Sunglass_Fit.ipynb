```python
import numpy as np
import cv2
```

```python
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
sunglasses = cv2.imread("birthday-cap-vector-illustration-31574811.webp",cv2.IMREAD_UNCHANGED) # IMREAD_UNCHANGED remove that background
sunglasses.shape   
cap =cv2.VideoCapture(0)
while True:
    flag,frame = cap.read()
    if(not flag):
        print("Camera not working")
        break
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray,1.3)
    for x,y,w,h in faces:
        #Resize the sunglasses according to face width
        #width = face width 
        # height (assuming) = 40% of face height
        overlay_width = w
        overlay_height = int(h*0.4)
        resized_sunglasses = cv2.resize(sunglasses,(overlay_width,overlay_height))
        x_offset = x
        y_offset = y+h//4
        
        overlay_img = resized_sunglasses[:,:,:3]
        mask = resized_sunglasses[:,:,3]
        #Create inverse mask (for background)
        mask_inv = cv2.bitwise_not(mask)

        #Define ROI - Region of Interest
        roi = frame[y_offset:y_offset+overlay_height,x_offset:x_offset+overlay_width]
        bg = cv2.bitwise_and(roi,roi,mask=mask_inv)
        fg = cv2.bitwise_and(overlay_img,overlay_img,mask=mask)
        combined=cv2.add(bg,fg)
        frame[y_offset:y_offset+overlay_height,x_offset:x_offset+overlay_width]=combined
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
        
    cv2.imshow("Beard filter",frame)
    key =cv2.waitKey(1) & 0xff
    if(key==27):
        break
cap.release()
cv2.destroyAllWindows()
```
